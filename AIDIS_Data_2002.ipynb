```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Standardization data

```python
from sklearn.preprocessing import StandardScaler
std_scale = StandardScaler()
X_train_std = std_scale.fit_transform(x_train)
X_test_std  = std_scale.fit_transform(x_test)
```

# Applying linear regression on whole data

```python
from sklearn import linear_model
import statsmodels.api as sm
regr = linear_model.LinearRegression()
model = regr.fit(X_train_std, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
y_predicted = model.predict(X_test_std)
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_predicted))
print("Absolute Mean squared error: %.2f" % mean_absolute_error(y_test, y_predicted))
print('R: %.2f' % r2_score(y_test, y_predicted))
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_predicted)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k-', lw=4)
ax.set_xlabel('Independent Variables')
ax.set_ylabel('Weight')
plt.show()
```
